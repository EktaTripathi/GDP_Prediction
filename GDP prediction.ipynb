## Model Development - Building the model with the right selection of the parameters

```python
factors=['Phones (per 1000)','Service','Birthrate','Agriculture','Literacy (%)','Net migration']
```

```python
X=df[factors]
y=df['GDP ($ per capita)']
```

```python
X
```

```python
y
```

## Model Evaluation - Testing the model on sufficient sample for correctness

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
len(X_train)
```

### LINEAR MODEL : LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
clf.fit(X_train,y_train)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.predict(X_test)
```

```python
y_test
```

```python
clf.score(X_test,y_test)
```

```python
training_data_accuracy=clf.score(X_train,y_train)
print("Training data accuracy for Linear regression= ",training_data_accuracy)
```

```python
testing_data_accuracy=clf.score(X_test,y_test)
print("Testing data accuracy for Linear regression= ",testing_data_accuracy)
```

### NON-LINEAR MODEL - RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model2=RandomForestRegressor()
model2.fit(X_train,y_train)
```

```python
training_data_accuracy=model2.score(X_train,y_train)
print("Training data accuracy for Random Forest Regression= ",training_data_accuracy)
```

```python
testing_data_accuracy=model2.score(X_test,y_test)
print("Testing data accuracy for Random Forest Regression= ",testing_data_accuracy)
```

Now, as we can see in the above few code cells, it is clear that non-linear model i.e., Random forest regressor is giving us a better accuracy than the linear regression model
